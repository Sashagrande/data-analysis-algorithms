```python
from sklearn.datasets import make_regression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from time import time
```

```python
dataset, target = make_regression(n_samples=100, n_features=100)
data = dataset.copy().T
data.shape
```

```python
normalization_data = lambda data : np.array([(arr - arr.min()) / (arr.max() - arr.min()) for arr in data])
data = normalization_data(data)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches((10, 5))
[ax.hist(data[np.random.randint(0, data.shape[0])]) for ax in axes.flat]
plt.show()
```

Видим что значения признаков колеблются в пределах от 0 до 1, значит нормализация прошла успешно

```python
calc_mse = lambda y, y_pred: np.mean((y - y_pred)**2)
```

```python
def eval_model(X, y, alpha, iterations, l=None):
    start_time = time()
    W = np.random.randn(X.shape[0])
    n = X.shape[1]
    errs = list()  # список с ошибками
    for i in range(1, iterations + 1):
        y_pred = np.dot(W, X)
        err = calc_mse(y, y_pred)
        errs.append(err)
        if l is None:
            W -= (alpha * (1/n * 2 * np.dot((y_pred - y), X.T)))
        else:
            W -= (alpha * (1/n * 2 * np.dot((y_pred - y), X.T))) + W * l

    return W, errs
```

```python
def eval_sgd_model(X, y, alpha,iterations, l=None, qty_in_batch=1):
    W = np.random.randn(X.shape[0])
    errs = list()
    n = X.shape[1]
    n_batch = n // qty_in_batch
    if n % qty_in_batch != 0:
        n_batch += 1
    for i in range(1, iterations+1):
        for b in range(n_batch):
            start_ = qty_in_batch*b
            end_ = qty_in_batch*(b+1)
            X_tmp, y_tmp = X[:, start_ : end_], y[start_ : end_]
            y_pred_tmp = np.dot(W, X_tmp)
            err = calc_mse(y_tmp, y_pred_tmp)
            if l is None:
                W -= (alpha * (1/n * 2 * np.dot((y_pred_tmp - y_tmp), X_tmp.T)))
            else:
                W -= (alpha * (1/n * 2 * np.dot((y_pred_tmp - y_tmp), X_tmp.T))) + W * l
        errs.append(err)
    return W, errs
```

```python
W1, errs1 = eval_model(data, target, 1e-2, 5000)
W2, errs2 = eval_sgd_model(data, target, 1e-2, 5000)
```

```python
plt.plot(range(len(errs1)), errs1, label='град спуск')
plt.plot(range(len(errs2)), errs2, label='стох. град спуск')
plt.xlabel('Iteration number')
plt.ylabel('MSE')
plt.title(f'MSE, 5000 итераций')
plt.legend()
```

## Вывод:
Генерил датасет наверно раз сто, пробовал разные размерности, получив графики mse, демонстрирущие отличия двух разных способов обучения модели на градиентном спуске решил остановиться. В этом случае стохастический градиентный спуск быстрее пришел к минимальному значению ошибки mse, но в остальных случаях бывали ситуации, когда они работали примерно одинаково.
